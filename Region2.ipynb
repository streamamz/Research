```python
import MDUS
from MDUS import Analysis
```

```python
MDUS.dip
```

```python
data = MDUS.MagData()
orbit = 201
data.Input(orbit=orbit,sec=10)
data.MoveAverage(window=13)
data.CalcKTH22(rc=False,ns=False)
ds = MDUS.dip.query('index == @orbit')['start'].values[0]
de = MDUS.dip.query('index == @orbit')['end'].values[0]
data.rotBCurrent(ds=ds,de=de)
```
