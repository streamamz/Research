```python
# マクスウェル分布を返す関数の作成
import numpy as np
import matplotlib.pyplot as plt

def maxwell_dist(v, T):
    k = 1.38e-23
    m = 1.67e-27
    return 4 * np.pi * (m / (2 * np.pi * k * T)) ** (3 / 2) * v ** 2 * np.exp(- m * v ** 2 / (2 * k * T))

# 三次元の速度空間で，
# maxwell_dist関数を使って温度が300Kの時の三次元プロットを作成
v = np.linspace(0, 2000, 1000)
T = 300
f = maxwell_dist(v, T)
plt.plot(v, f)
plt.xlabel('v [m/s]')
plt.ylabel('f(v) [s/m]')
plt.show()

```
